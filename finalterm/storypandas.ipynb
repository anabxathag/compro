1. Pandas data structures

```python
# create pandas series data
students = pd.Series([62070229, 62070230, 62070231, 62070232, 62070233,
            62070234, 62070235, 62070236, 62070237, 62070238])
print(students)
print("-" * 20)
# access value: series_name[index]
print(students[0])
```

```python
# create pandas series data with index - 1
students = pd.Series([62070229, 62070230, 62070231, 62070232, 62070233]
                     , index = [2, 4, 6, 8, 10])
print(students)
```

```python
# create pandas series data with index - 2
students = pd.Series({2 : 62070229, 4 : 62070230, 6: 62070231, 8 : 62070232, 10 : 62070233})
print(students)
print("-" * 20)
# get series's index value
print(students.index)
```

4. การคำนวณและจัดการข้อมูลตัวเลข

```python
# loading a .csv file into a pandas DataFrame
zoo = pd.read_csv('zoo.csv', delimiter = ',')
zoo
```

```python
# get the DataFrame shape (rows, columns)
rows, cols = zoo.shape
print(rows, cols)
print("-" * 20)
# et the DataFrame #rows, #columns
print(len(zoo), len(zoo.columns))
```

```python
# count the number of row in DataFrame
print(zoo.count())
print("-" * 20)
print(zoo[['animal']].count())
print("-" * 20)
print(zoo.animal.count())
print("-" * 20)
# count the number of tiger
print(len(zoo[zoo.animal == 'tiger']))
```

```python
# example of groupby()
print(zoo.groupby('animal').size())
print("-" * 40)
print(zoo.groupby('animal').count())
print("-" * 40)
# the summation of water_need of each animal category
print(zoo.groupby('animal').water_need.sum())
```

```python
# find the mininum water_need
print(zoo.water_need.min())
print("-" * 40)
# find the maximum water_need
print(zoo.water_need.max())
print("-" * 40)
# find the water_need of all elephant
print(zoo[zoo.animal == 'elephant'].water_need)
print("-" * 40)
# find the maximum water_need of elephant
print(zoo[zoo.animal == 'elephant'].water_need.max())
```

```python
# Creating the DataFrame
tmp = pd.DataFrame({"A":[43, 56, 12, 36],
                "B":[11, 25, 90, 81],
                "C":[61, 18, 66, 47]})
tmp
```

```python
# applying min() function: row-wise
print(tmp.min(axis=0))
print("-" * 40)
# applying idxmin() function: row-wise
print(tmp.idxmin(axis=0))
print("-" * 40)
# applying min() function: column-wise
print(tmp.min(axis=1))
print("-" * 40)
# applying idxmin() function: column-wise
print(tmp.idxmin(axis=1))
```

```python
samples = pd.Series([9, 3, 1, 8, 3, 6])
# display row data
samples
```

```python
# mean value: sum(samples) / len(samples)
print('mean:', samples.mean())
print("-" * 25)
# median value: sort(samples), then pick the value in the middle
print('median:', samples.median())
print(zoo.water_need.median())  # water_need: median
print("-" * 25)
# mode value: most frequency values
print('mode:', samples.mode())
print(zoo.water_need.mode())  # water_need: mod
print("-" * 25)
# standard deviation: the amount of dispersion of a set of values
print('sd:', samples.std())
print(zoo.water_need.std())  # water_need: sd
```
