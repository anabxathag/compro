```python
# loading a .csv file into a pandas DataFrame
records = pd.read_csv('pandas_tutorial_read.csv', delimiter = ';',
        names = ['my_datetime', 'event', 'country', 'user_id', 'source', 'topic'])
tutor = records.copy()
tutor
```

# แบบฝึกหัดที่ 1

1. จงแสดงข้อมูล แถวเลขคี่ ระหว่างแถวที่ 25-50 ที่ประกอบด้วย 'user_id', 'source' และ 'topic'

```python
tutor.loc[25:50:2, ['user_id', 'source', 'topic']]
```

2. จงแสดงข้อมูลผู้ใช้ที่มาจากประเทศ 'country_2' หรือ 'country_6'

```python
df_1 = records[records.country == 'country_2']
df_2 = records[records.country == 'country_6']
ans = pd.concat([df_1, df_2], axis=0)
ans.sort_index(ascending=True)
```

3. จงแสดงข้อมูลผูใช้ที่มาจากแหล่ง 'SEO' และหัวข้อ'North America' จำนวน10คนใด (แบบสุ่มๆ)

```python
tmp = tutor[(tutor.source == 'SEO') & (tutor.topic == 'North America')]
tmp.sample(10)
```

# แบบฝึกหัดที่ 2

```python
# loading a .csv file into a pandas DataFrame
zoo = pd.read_csv('zoo_updated.csv', delimiter = ',')
suiya = zoo.copy()
suiya
```

1. จงแสดงสัตว์ทุกตัวที่มีอายุเกิน 5 ปีและต้องการน้ำมากกว่า 400 หน่วย

```python
suiya[(suiya.age > 5) & (suiya.water_need > 400)]
```

2. จงแบ่งสัตว์ออกตามอายุ และประเภท พร้อมทั้งแสดงจำนวนสัตว์ในกลุ่ม

```python

suiya.groupby(["animal", "age"])["age"].count()
```

3. จงแสดงประเภทสัตว์ที่มีจำนวนเยอะที่สุดในสวนสัตว์ และระบุจำนวนดังกล่าว

```python
ani_coun = suiya["animal"].value_counts()
ani_coun[ani_coun == max(ani_coun)]
```

# แบบฝึกหัดที่ 3

```python
suiya
```

1. จงแสดงข้อมูลสัตว์ที่ต้องการปริมาณน้ำที่สูงที่สุดในแต่ละประเภท เรียงลำดับตามอายุจากมากไปหาน้อย

```python
suiya.groupby('animal').max('water_need').sort_values(["age"], ascending=0)
```

2. จงแสดงข้อมูลสัตว์เรียงลำดับตามอายุจากน้อยไปหามาก และเรียงตามน้ำหนักจากมากไปหาน้อย

```python
suiya.sort_values(['animal', 'age', 'weight'], ascending=[1, 1, 0])
```

3. จงแสดงค่าปริมาณน้ำที่สัตว์แต่ละประเภทต้องการสูงสุด 3 อันดับแรก
(หมายเหตุ: คำสั่ง .nlargest(n) ใช้เพื่อแสดงข้อมูลสูงที่สุด n ลำดับ)

```python
suiya.groupby('animal')['water_need'].nlargest(3)
```
